# Relatório de análise 

### Criando novas variáveis 

```python
import pandas as pd
```

```python
dados = pd.read_csv('dados_residencial_final_v2.csv', sep=';')
```

```python
dados.head(10)
```

### para adicionar uma nova variável ao DF

```python
dados['Valor Bruto'] = dados['Valor'] + dados['Condominio'] + dados['IPTU']
```

```python
dados.head(10)
```

### nova variável m2: 

```python
dados['Valor por m2'] = dados['Valor'] / dados['Area']
```

```python
dados.head(10)
```

### para reduzir para duas casas decimais: 

```python
dados['Valor por m2'] = dados['Valor por m2'].round(2)
```

```python
dados.head(10)
```

### adicionando valor m2 bruto

```python
dados['Valor por m2/Bruto'] = (dados['Valor Bruto'] / dados['Area']).round(2)
```

```python
dados.sample(10)
```

### agregando por tipo: 

```python
tipos = dados['Tipo']
tipos.drop_duplicates()
```

```python
# criando primeiro uma lista com os valores desejados:
casa = ['Casa', 'Casa de Condomínio', 'Casa de Vila','Sítio']
```

```python
# adicionando nova tabela ao dataFrame
```

```python
dados['Tipo Agregado'] = dados['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Apartamento')
```

```python
dados.sample(10)
```

# Excluindo variáveis 

```python
# criando um dataframe com as variáveis desejadas:
dados_aux = pd.DataFrame(dados[['Tipo Agregado', 'Valor por m2','Valor Bruto','Valor por m2/Bruto']])
```

```python
dados_aux.head(10)
```

```python
# deletando uma coluna inteira:
del dados_aux['Valor Bruto']
```

```python
dados_aux.head(10)
```

```python
# deletando uma coluna usando .pop()
dados_aux.pop('Valor por m2/Bruto')
```

```python
dados_aux.head(10)
```

```python
# excluindo variáveis usando .drop()
```

```python
dados.drop(['Valor Bruto', 'Valor por m2/Bruto'], axis = 1, inplace = True)
# o axis informa a direção axis = 1 => coluna axis = 0 => linha
```

```python
dados.head(10)
```

```python
# exportando banco de dados atualizado:

dados.to_csv('dados_residencial_final_v3.csv', sep=';', index = False)
```
