### Seleções e Frequencias

```python
import pandas as pd
```

```python
dados = pd.read_csv('dados_residencial_final.csv', sep = ';')
```

```python
dados.head(10)
```

# Selecione somente os imóveis classificados com tipo 'Apartamento'.

```python
selecao_apartamento = dados['Tipo'] == 'Apartamento'
```

```python
apartamento = dados[selecao_apartamento]
```

```python
apartamento.head(10)
```

```python
apartamento.shape[0]
```

## Selecione os imóveis classificados com tipos 'Casa', 'Casa de Condomínio' e 'Casa de Vila'.

```python
selecao_casas = (dados['Tipo'] == 'Casa') | (dados['Tipo'] == 'Casa de Condomínio') | (dados['Tipo'] == 'Casa de Vila') 
```

```python
selecao_casas = dados[selecao_casas]
```

```python
selecao_casas.head(10)
```

```python
selecao_casas.shape[0]
```

## Selecione os imóveis com área entre 60 e 100 metros quadrados, incluindo os limites.

```python
selecao_area = (dados['Area'] >= 60) & (dados['Area'] <= 100)
```

```python
selecao_area
```

```python
selecao_area = dados[selecao_area]
selecao_area.head(10)
```

```python
selecao_area.shape[0]
```

## Selecione os imóveis que tenham pelo menos 4 quartos e aluguel menor que R$ 2.000,00.

```python
selecao_quartos = (dados['Quartos'] >= 4) & (dados['Valor'] < 2000)
```

```python
selecao_quartos = dados[selecao_quartos]
selecao_quartos.head(10)
```

```python
selecao_quartos.shape[0]
```
