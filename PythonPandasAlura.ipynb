## Separando imóveis residênciais

```python
### criando lista usando .drop
```

```python
list(dados_residencial2['Tipo'].drop_duplicates())
```

```python
residencial = ['Quitinete',
 'Casa',
 'Apartamento',
 'Casa de Condomínio',
 'Flat',
 'Casa de Vila',
 'Sítio']
```

```python
residencial 
```

```python
### usando isin() para definir as contruções residenciais 
```

```python
selecao = dados_residencial2['Tipo'].isin(residencial)
```

```python
selecao
```

```python
### retirando os valores False do dataframe original
```

```python
dados_residencial_final = dados_residencial2[selecao]
```

```python
### DataFrame final só com valores Tipo residencial 
```

```python
dados_residencial_final.sample(10)
```

```python
### verificação final
```

```python
list(dados_residencial_final['Tipo'].drop_duplicates())
```

```python
### comparação 
```

```python
dados_residencial2.shape[0]
```

```python
dados_residencial_final.shape[0]
```

```python
dados_residencial2.shape[0] - dados_residencial_final.shape[0]
```

```python
### reconstruindo o index
```

```python
dados_residencial_final.index = range(dados_residencial_final.shape[0])
```
