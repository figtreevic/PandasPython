```python
import pandas as pd
```

```python
dados = pd.read_csv('dados_residencial_final_v3.csv', sep=';')
```

```python
dados.sample(10)
```

```python
# valor média geral:

dados['Valor'].mean()
```

```python
bairros = dados['Bairro'].unique()
```

```python
selecao = dados['Bairro'].isin(bairros)
```

```python
dados_bairros = dados[selecao]
```

```python
dados_bairros.head(10)
```

```python
# usando .groupby()

dados_bairros = dados.groupby('Bairro')
```

```python
# o resultado é uma biblioteca que armazaena os indexes pelo parâmetro informado:

type(dados_bairros)
```

```python
#verificando:
dados_bairros.groups
```

```python
# analisando os dados:

for bairro, data in dados_bairros:
    print(bairro)
```

```python
for bairro, dados in dados_bairros:
    print(data)
```

```python
# calculando valor médio por bairro:

for bairro, data in dados_bairros:
    print('{} -> {}'.format(bairro, data['Valor'].mean()))
```

```python
# maneira mais simples de executar a tarefa acima:

dados_bairros[['Valor']].mean().round(2).sample(10)
```

# Estatísticas descritivas

```python
# descriçao detalhada do dados_bairros usando .describe()
dados_bairros.head(10)
```

```python
dados_bairros['Valor'].describe().round(2)
```

```python
# entendendo o dataframe criado com .describe():
# count: frequencia
# mean: média
# std: desvio padrão
# min: valor mínimo
# 25%: primeiro-quartil => 25% dos dados estão abaixo deste valor
# 50%: segundo-quartil => 50% dos dados estão abaixo deste valor
# 75%: terceiro-quartil => 75% dos dados estão abaixo deste valor
# max: valor máximo
```

```python
# usando.aggregate() e escolhendo parâmetros:
dados_bairros['Valor'].aggregate(['min', 'max', 'sum'])
```

```python
# usando .rename() para nomear colunas:

dados_bairros['Valor'].aggregate(['min','max']).rename(columns = {'min': 'Mínimo', 'max': 'Máximo'})
```

# Utilizando gráricos para visualização de dados: 

```python
%matplotlib inline 
#confg. necessária Jupyter. Mostra gráfico estático
import matplotlib.pyplot as plt
plt.rc('figure', figsize = (20, 10)) 
#configurando tamanho do gráfico na tela
```

```python
# verificando desvio padrão usando gráficos:
dados_bairros['Valor'].std().plot.bar(color ='blue')
# desvio padrao .std() .plot.bar() gráfico de barra
```

```python
# verificando média usando gráficos:
fig2 = dados_bairros['Valor'].mean().plot.bar(color = 'blue')
fig2.set_ylabel('Valor do Aluguel')
#configurando título para o eixo y
fig2.set_title('Valor médio do Aluguel / Bairro', {'fontsize': 22})
# configurando título gráfico e tamanho de fonte               
```
