## Tratamento da coluna Condomínio

#### para descobrir os valores nulos dentro do field

```python
dados[dados['Condominio'].isnull()].shape[0]
```

#### verificando quais desses são em apartamentos: 

```python
selecao = (dados['Tipo'] == 'Apartamento') & (dados['Condominio'].isnull())
```

##### Contudo, se nos atentarmos a nossa seleção, verificaremos que ela nos fornece exatamente as assinaturas que deverão ser removidas do banco de dados. Reaproveitaremos a estrutura já utilizada anteriormente neste notebook, e passaremos a seleção que criamos (selecao). Para eliminar as assinaturas nulas selecionadas ao invés de coletá-las, iremos adicionar ~ em selecao, que inverte a Series booleana. 

```python
A = dados.shape[0] #matriz original
dados = dados[~selecao] #removendo os valores selecionados anterioremente
B = dados.shape[0]
A - B
```

```python
dados[dados['Condominio'].isnull()].shape[0]
```

###### os valores restantes de 1121 são aqueles que tem condomínio nulo, mas não são apartamentos

```python
dados.fillna(0, inplace = True)
```

###### outra maneira de fazer a mesma substituição

```python
dados = dados.fillna({'Condominio': 0, 'IPTU': 0}) # aqui é mais favorável para definir valores diferentes
```
